Question 1: What is hypothesis testing in statistics?   
Answer- Hypothesis testing is a statistical method used to make inferences about a population based on sample data. It's a formal procedure for determining whether a specific statement or claim about a population parameter (like a mean or a proportion) is likely to be true. Essentially, it helps you decide if the results you've observed in a sample are statistically significant or if they could have occurred by random chance.

 Question 3: Explain the significance level in hypothesis testing and its role in deciding
the outcome of a test.    
Answer:The significance level, denoted by the Greek letter alpha (Î±), is a crucial threshold you set before starting a hypothesis test. It represents the maximum probability of making a Type I errorâ€”which is the error of incorrectly rejecting the null hypothesis when it's actually true.


Role in Deciding the Outcome
The significance level acts as the benchmark for your decision. After you perform your statistical test, you calculate a p-value. The p-value is the probability of observing your data (or something more extreme) if the null hypothesis were true.



The decision rule is simple:

If p-value â‰¤Î±: You reject the null hypothesis. The observed result is considered "statistically significant" because the probability of it occurring by random chance (if the null were true) is lowâ€”at or below your predetermined significance level.


If p-value >Î±: You fail to reject the null hypothesis. The result is not statistically significant because the data you observed is not unusual enough to dismiss the possibility that it occurred by chance.

Common Significance Levels
The most common significance level is Î±=0.05. This means there's a 5% chance of rejecting a true null hypothesis. Other common values include 0.01 (a stricter test, 1% risk) and 0.10 (a more lenient test, 10% risk). The choice of Î± depends on the context of your study and the consequences of making a Type I error. For example, in medical trials, a lower Î± might be used to minimize the risk of approving an ineffective drug.

Question 5: What is the difference between a Z-test and a T-test? Explain when to use each.    
Answer- A Z-test and a T-test are both types of hypothesis tests used to determine if a sample mean is significantly different from a population mean. The key difference lies in what you know about the population and the size of your sample.


Z-test ðŸ§ª
A Z-test is used when you know the population standard deviation (Ïƒ) or when you have a large sample size (generally nâ‰¥30). According to the Central Limit Theorem, for large samples, the distribution of the sample means will be approximately normal, so a Z-test can be used even if you don't know the population standard deviation, as the sample standard deviation will be a reliable estimate.


When to use a Z-test:

The population standard deviation is known. This is rare in practice.

The sample size is large (nâ‰¥30).

T-test ðŸ”¬
A T-test is used when the population standard deviation is unknown and you are working with a small sample size (generally n<30). Since you're using the sample standard deviation (s) as an estimate for the population standard deviation, there's more uncertainty. The T-test uses a t-distribution, which has "fatter tails" than the normal distribution to account for this extra uncertainty. As the sample size increases, the t-distribution approaches the normal distribution, and the results of a T-test and a Z-test become very similar.


When to use a T-test:

The population standard deviation is unknown.

The sample size is small (n<30).

The choice between the two tests is a critical first step in hypothesis testing, as it ensures you are using the correct statistical distribution for your data.   

Question 6: Write a Python program to generate a binomial distribution with n=10 and
p=0.5, then plot its histogram.

Answer-



```python
import numpy as np
import matplotlib.pyplot as plt

def plot_binomial_distribution(n, p, size=1000):
    """
    Generates a binomial distribution and plots a histogram.

    Args:
        n (int): The number of trials.
        p (float): The probability of success in each trial.
        size (int): The number of random samples to generate.
    """
    # 1. Generate the random samples from a binomial distribution
    # The numpy.random.binomial function is used for this purpose.
    # It takes the number of trials (n), the probability of success (p),
    # and the number of samples to generate (size) as arguments.
    random_samples = np.random.binomial(n, p, size)

    # 2. Create the histogram
    # The plt.hist function plots the histogram.
    # The `bins` argument ensures that each possible outcome (0 to 10)
    # has its own bar. `rwidth` controls the width of the bars.
    plt.hist(random_samples, bins=np.arange(-0.5, n + 1.5, 1), rwidth=0.8, color='skyblue', edgecolor='black')

    # 3. Add labels and a title for better readability
    plt.title(f'Binomial Distribution (n={n}, p={p})')
    plt.xlabel('Number of Successes')
    plt.ylabel('Frequency')

    # 4. Set x-axis ticks to show the discrete outcomes
    plt.xticks(np.arange(0, n + 1, 1))

    # 5. Display the grid
    plt.grid(axis='y', alpha=0.75)

    # 6. Show the plot
    plt.show()

# Define the parameters for the binomial distribution
n_trials = 10
p_success = 0.5

# Call the function to generate and plot the distribution
plot_binomial_distribution(n_trials, p_success)
```

Question 7: Implement hypothesis testing using Z-statistics for a sample dataset in
Python. Show the Python code and interpret the results.
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
50.3, 50.4, 50.0, 49.7, 50.5, 49.9]   

Answer-

```python
import numpy as np
from scipy import stats

def perform_z_test(sample_data, pop_mean, pop_std, alpha=0.05):
    """
    Performs a one-sample Z-test on a given dataset.

    Args:
        sample_data (list): The list of numerical data for the sample.
        pop_mean (float): The hypothesized population mean (mu_0).
        pop_std (float): The known population standard deviation (sigma_0).
        alpha (float): The significance level, default is 0.05.
    """
    # 1. Define the null and alternative hypotheses
    # H0 (Null Hypothesis): The sample mean is equal to the population mean.
    # H1 (Alternative Hypothesis): The sample mean is not equal to the population mean.
    print(f"Null Hypothesis (H0): The population mean is equal to {pop_mean}")
    print(f"Alternative Hypothesis (H1): The population mean is not equal to {pop_mean}\n")

    # 2. Calculate key statistics from the sample data
    sample_size = len(sample_data)
    sample_mean = np.mean(sample_data)

    print(f"Sample Size (n): {sample_size}")
    print(f"Sample Mean (x_bar): {sample_mean:.4f}")
    print(f"Known Population Standard Deviation (sigma): {pop_std}\n")

    # 3. Calculate the Z-statistic
    # Formula: Z = (sample_mean - pop_mean) / (pop_std / sqrt(sample_size))
    z_statistic = (sample_mean - pop_mean) / (pop_std / np.sqrt(sample_size))

    # 4. Calculate the two-tailed p-value
    # The stats.norm.sf() function gives the survival function (1 - CDF),
    # so we multiply by 2 for a two-tailed test.
    p_value = 2 * stats.norm.sf(abs(z_statistic))

    print(f"Calculated Z-statistic: {z_statistic:.4f}")
    print(f"P-value: {p_value:.4f}\n")

    # 5. Make a decision based on the p-value and alpha
    print(f"Significance Level (alpha): {alpha}")
    if p_value < alpha:
        print(f"Since the p-value ({p_value:.4f}) is less than alpha ({alpha}), we reject the null hypothesis.")
        print("Conclusion: The sample mean is significantly different from the population mean.")
    else:
        print(f"Since the p-value ({p_value:.4f}) is greater than or equal to alpha ({alpha}), we fail to reject the null hypothesis.")
        print("Conclusion: There is not enough evidence to say the sample mean is significantly different from the population mean.")

# Provided sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Assume a population mean and a known population standard deviation
# For a valid Z-test, these must be known. We'll use reasonable assumptions.
assumed_pop_mean = 50.0
known_pop_std = 1.0

# Call the function to run the Z-test and print the results
perform_z_test(sample_data, assumed_pop_mean, known_pop_std)
```

Question 8: Write a Python script to simulate data from a normal distribution and
calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.   

Answer-

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def calculate_confidence_interval(data, confidence_level=0.95):
    """
    Calculates the confidence interval for the mean of a dataset.

    Args:
        data (np.array or list): The dataset.
        confidence_level (float): The desired confidence level (e.g., 0.95 for 95%).

    Returns:
        tuple: A tuple containing the mean, margin of error, and the lower and upper bounds of the confidence interval.
    """
    # Convert data to a numpy array for easier calculations
    data = np.array(data)

    # Calculate sample statistics
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)  # Use ddof=1 for sample standard deviation
    sample_size = len(data)

    # Calculate the t-critical value for the given confidence level and degrees of freedom
    # Degrees of freedom (df) = n - 1
    df = sample_size - 1
    t_critical = stats.t.ppf((1 + confidence_level) / 2, df)

    # Calculate the standard error of the mean (SEM)
    sem = sample_std / np.sqrt(sample_size)

    # Calculate the margin of error
    margin_of_error = t_critical * sem

    # Calculate the confidence interval bounds
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return sample_mean, margin_of_error, lower_bound, upper_bound

def simulate_and_plot(mean, std_dev, sample_size, confidence_level=0.95):
    """
    Simulates data from a normal distribution, calculates the confidence interval,
    and plots a histogram of the data with the confidence interval.

    Args:
        mean (float): The mean of the normal distribution to simulate.
        std_dev (float): The standard deviation of the normal distribution to simulate.
        sample_size (int): The number of data points to simulate.
        confidence_level (float): The confidence level for the interval.
    """
    # 1. Simulate data from a normal distribution
    np.random.seed(42)  # for reproducibility
    simulated_data = np.random.normal(loc=mean, scale=std_dev, size=sample_size)

    # 2. Calculate the confidence interval
    sample_mean, margin_of_error, lower_bound, upper_bound = calculate_confidence_interval(
        simulated_data, confidence_level
    )

    # 3. Print the results to the console
    print(f"Simulated Sample Size: {sample_size}")
    print(f"Simulated Sample Mean: {sample_mean:.4f}")
    print(f"Confidence Level: {confidence_level * 100}%")
    print(f"Margin of Error: Â±{margin_of_error:.4f}")
    print(f"Confidence Interval: [{lower_bound:.4f}, {upper_bound:.4f}]")

    # 4. Plot the data
    plt.figure(figsize=(10, 6))

    # Plot the histogram of the simulated data
    plt.hist(simulated_data, bins=30, edgecolor='black', alpha=0.7, label='Simulated Data')

    # Plot vertical lines for the confidence interval and sample mean
    plt.axvline(sample_mean, color='red', linestyle='--', linewidth=2, label='Sample Mean')
    plt.axvline(lower_bound, color='green', linestyle='-', linewidth=2, label='Confidence Interval')
    plt.axvline(upper_bound, color='green', linestyle='-', linewidth=2)

    # Add labels and a title
    plt.title(f'Simulated Normal Distribution with {confidence_level*100}% Confidence Interval')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()

# Set simulation parameters
pop_mean = 100
pop_std = 15
sim_sample_size = 50

# Run the simulation and plot the results
simulate_and_plot(pop_mean, pop_std, sim_sample_size)
```

Question 9: Write a Python function to calculate the Z-scores from a dataset and
visualize the standardized data using a histogram. Explain what the Z-scores represent
in terms of standard deviations from the mean.

Answer-

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_zscores(data):
    """
    Calculates the Z-scores for a dataset and plots a histogram
    of the standardized data.

    Args:
        data (list or np.array): The dataset to standardize.
    """
    # 1. Convert the data to a NumPy array for efficient calculations
    data = np.array(data)

    # 2. Calculate the sample mean and standard deviation
    # We use ddof=0 for population standard deviation to match the common
    # Z-score formula, assuming the input data is the entire population
    # for the purpose of this standardization exercise.
    mean = np.mean(data)
    std_dev = np.std(data, ddof=0)

    # Handle case where standard deviation is zero to avoid division by zero
    if std_dev == 0:
        print("Standard deviation is zero, cannot calculate Z-scores.")
        return

    # 3. Calculate the Z-score for each data point
    # Formula: Z = (x - mean) / std_dev
    z_scores = (data - mean) / std_dev

    # 4. Print key statistics of the original and standardized data
    print("Original Data Statistics:")
    print(f"  Mean: {mean:.4f}")
    print(f"  Standard Deviation: {std_dev:.4f}\n")

    print("Standardized Data (Z-scores) Statistics:")
    print(f"  Mean: {np.mean(z_scores):.4f}")
    print(f"  Standard Deviation: {np.std(z_scores, ddof=0):.4f}\n")

    # 5. Plot the histogram of the Z-scores
    plt.figure(figsize=(10, 6))

    # Create the histogram
    plt.hist(z_scores, bins=15, edgecolor='black', alpha=0.7)

    # Add a vertical dashed line at the mean (Z-score = 0)
    plt.axvline(0, color='red', linestyle='--', linewidth=2, label='Mean (Z=0)')

    # Add labels and a title
    plt.title('Histogram of Z-scores (Standardized Data)')
    plt.xlabel('Z-score (Standard Deviations from the Mean)')
    plt.ylabel('Frequency')
    plt.legend()
    plt.grid(axis='y', linestyle='--', alpha=0.6)

    # Show the plot
    plt.show()

# Sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Run the function with the sample data
calculate_and_plot_zscores(sample_data)
```
